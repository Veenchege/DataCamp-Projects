```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://www.jumia.co.ke/'
product_links = []

for x in range(1, 51):
    r = requests.get(f'https://www.jumia.co.ke/catalog/?q=laptops&page={x}#catalog-listing')
    soup = BeautifulSoup(r.content, 'html.parser')
    links = soup.find_all('a', class_='core')

    for link in links:
        href = link.get('href')
        if href is not None:
            product_links.append(base_url + href)

# print all product links
for product_link in product_links:
    print(product_link)
```

```python
product_list = []

for links in product_links:
  r = requests.get(links)
  soup = BeautifulSoup(r.content, 'html.parser')
  try:
    Name = soup.find('h1', class_='-fs20 -pts -pbxs').text
  except:
    pass
  try:
    Current_price = pd.to_numeric(soup.find('span', class_='-b -ltr -tal -fs24 -prxs').text.strip('KSh ').replace(',', ''))
  except:
    pass
  try:
    Original_price =  pd.to_numeric(soup.find('span', class_='-tal -gy5 -lthr -fs16 -pvxs').text.strip('KSh ').replace(',', ''))
  except:
    pass
  try:
    Discount = soup.find('span', class_='bdg _dsct _dyn -mls').text
  except:
    pass
  try:
    Brand = soup.find_all('a', class_='_more')[1].text
  except:
    pass
  try:
    Store_name = soup.find('p', class_='-m -pbs').text
  except:
    pass
  try:
    Seller_score = soup.find('bdo', class_='-m -prxs').text
  except:
    pass
  try:
    Warranty = soup.find_all('div', class_='markup -ptxs')[4].text
  except:
    pass


  jumiadata = {
      'Product_name' : Name,
      'Product_brand' : Brand,
      'Sale_price' : Current_price,
      'Original_price' : Original_price,
      'Discount' : Discount,
      'Store_name' : Store_name,
      'Seller_score' : Seller_score,
      'Warranty' : Warranty
  }

  product_list.append(jumiadata)
  print('Saving: ', jumiadata['Product_name'])

df = pd.DataFrame(product_list)
df.to_excel('jumiadata.xlsx')
print(df.head(15))
```
